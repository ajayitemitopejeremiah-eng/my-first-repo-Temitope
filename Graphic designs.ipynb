```python
import reportlab
from reportlab.lib.pagesizes import letter
from reportlab.lib.units import inch
from reportlab.lib import colors
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer, Table, TableStyle
from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle
from reportlab.lib.enums import TA_CENTER, TA_JUSTIFY

def create_tract(filename="tract_designpp.pdf"):
    # 1. Setup Document Geometry (Standard Letter size, clean margins)
    doc = SimpleDocTemplate(
        filename,
        pagesize=letter,
        rightMargin=0.5 * inch,
        leftMargin=0.5 * inch,
        topMargin=0.5 * inch,
        bottomMargin=0.5 * inch
    )
    
    story = []
    
    # 2. Define Clean Typography Styles
    styles = getSampleStyleSheet()
    
    # Custom Title Style
    title_style = ParagraphStyle(
        'TractTitle',
        parent=styles['Heading1'],
        fontName='Helvetica-Bold',
        fontSize=24,
        leading=28,
        textColor=colors.HexColor("#1A365D"), # Deep structural blue
        alignment=TA_CENTER,
        spaceAfter=15
    )
    
    # Custom Body Text Style
    body_style = ParagraphStyle(
        'TractBody',
        parent=styles['Normal'],
        fontName='Helvetica',
        fontSize=10,
        leading=14,
        textColor=colors.HexColor("#2D3748"), # Dark slate for readability
        alignment=TA_JUSTIFY,
        spaceAfter=10
    )
    
    # 3. Build Content Panels
    # Front Cover Content / Header
    story.append(Spacer(1, 0.5 * inch))
    story.append(Paragraph("IF NOT WHY NOT", title_style))
    story.append(Spacer(1, 0.2 * inch))
    
    # A thin accent divider line using a table structure
    divider = Table([[""]], colWidths=[7.5 * inch], rowHeights=[2])
    divider.setStyle(TableStyle([
        ('BACKGROUND', (0,0), (-1,-1), colors.HexColor("#1A365D")),
        ('TOPPADDING', (0,0), (-1,-1), 0),
        ('BOTTOMPADDING', (0,0), (-1,-1), 0),
    ]))
    story.append(divider)
    story.append(Spacer(1, 0.3 * inch))
    
    # Core Message / Content Paragraphs
    paragraph_1 = """
    IF NOT, WHY NOT?

Do you know that Someone really cares about you, as matter of fact the price for your soul was His life and that Someone is Jesus Christ.

Jesus Christ is the Son of God who took the form of man so as to give Himself for man.

The Bible tells us in John 15:13
Greater love hath no man than this that a man lay down His life for His friends.

The question to you today is have you made Him your friend? if not, why not?

 You can take this simple steps to make Him your friend;

A- Acknowledge that you are a sinner (Romans 3:23).

B- Believe that Jesus died to save you from sin (Romans 5:8).

C- Confess your sins ( 1John 1:9).

D- Depart from sin and determine never to return to sin ( 2 Corinthians 6:17).

Sinners Prayers;
Faithful Father lover of my soul, I thank You for sending Jesus Christ Your only Son to give Himself for me, I believe that He came to bring me back to You and I surrender my sins today in exchange for a new life in You. I accept Him as my personal Lord and Saviour, thank You Father for forgiving me of my sins. This I pray in Jesus name amen.

Congratulations if you said this prayers of faith, welcome to the family of God..
    """
    story.append(Paragraph(paragraph_1, body_style))
    
    paragraph_2 = """
   God bless you.
    """
    story.append(Paragraph(paragraph_2, body_style))
    
    # 4. Generate the Final PDF
    doc.build(story)

# Direct function execution to bypass all underscore errors
create_tract()
print("Tract PDF layout generated successfully.")
```

```python
import reportlab
from reportlab.lib.pagesizes import letter
from reportlab.lib.units import inch
from reportlab.lib import colors
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer, Table, TableStyle
from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle
from reportlab.lib.enums import TA_CENTER, TA_JUSTIFY, TA_LEFT
from reportlab.graphics.shapes import Drawing, Rect, Line, Circle

def create_attractive_tract(filename="enhanced_tract.pdf"):
    # Setup document with clean margins
    doc = SimpleDocTemplate(
        filename,
        pagesize=letter,
        rightMargin=0.6 * inch,
        leftMargin=0.6 * inch,
        topMargin=0.6 * inch,
        bottomMargin=0.6 * inch
    )
    
    story = []
    styles = getSampleStyleSheet()
    
    # --- Custom Color Palette ---
    PRIMARY_COLOR = colors.HexColor("#1A365D")  # Deep Regal Navy
    SECONDARY_COLOR = colors.HexColor("#C53030") # Deep Crimson Accent
    TEXT_COLOR = colors.HexColor("#2D3748")      # Charcoal Body Text
    BG_LIGHT = colors.HexColor("#EDF2F7")        # Soft Grey/Blue Box background
    
    # --- Custom Typography ---
    title_style = ParagraphStyle(
        'TractTitle',
        parent=styles['Heading1'],
        fontName='Helvetica-Bold',
        fontSize=28,
        leading=32,
        textColor=PRIMARY_COLOR,
        alignment=TA_CENTER,
        spaceAfter=4
    )
    
    subtitle_style = ParagraphStyle(
        'TractSubtitle',
        parent=styles['Normal'],
        fontName='Helvetica-BoldOblique',
        fontSize=12,
        leading=16,
        textColor=SECONDARY_COLOR,
        alignment=TA_CENTER,
        spaceAfter=15
    )
    
    body_style = ParagraphStyle(
        'TractBody',
        parent=styles['Normal'],
        fontName='Helvetica',
        fontSize=10.5,
        leading=15,
        textColor=TEXT_COLOR,
        alignment=TA_JUSTIFY,
        spaceAfter=12
    )
    
    step_style = ParagraphStyle(
        'TractSteps',
        parent=styles['Normal'],
        fontName='Helvetica-Bold',
        fontSize=11,
        leading=16,
        textColor=PRIMARY_COLOR,
        alignment=TA_LEFT,
        spaceAfter=6,
        leftIndent=15
    )
    
    prayer_title = ParagraphStyle(
        'PrayerTitle',
        parent=styles['Normal'],
        fontName='Helvetica-Bold',
        fontSize=12,
        leading=14,
        textColor=SECONDARY_COLOR,
        alignment=TA_CENTER,
        spaceAfter=6
    )
    
    prayer_body = ParagraphStyle(
        'PrayerBody',
        parent=styles['Normal'],
        fontName='Helvetica-Oblique',
        fontSize=10,
        leading=14,
        textColor=TEXT_COLOR,
        alignment=TA_CENTER
    )
    
    footer_style = ParagraphStyle(
        'TractFooter',
        parent=styles['Normal'],
        fontName='Helvetica-Bold',
        fontSize=12,
        leading=16,
        textColor=PRIMARY_COLOR,
        alignment=TA_CENTER,
        spaceBefore=15
    )

    # --- 1. Header Illustration (Vector Cross) ---
    d = Drawing(7.3 * inch, 40)
    d.add(Line(0, 20, 7.3 * inch, 20, strokeColor=PRIMARY_COLOR, strokeWidth=1))
    d.add(Circle(3.65 * inch, 20, 12, fillColor=colors.white, strokeColor=SECONDARY_COLOR, strokeWidth=2))
    d.add(Line(3.65 * inch, 12, 3.65 * inch, 28, strokeColor=PRIMARY_COLOR, strokeWidth=2))
    d.add(Line(3.55 * inch, 22, 3.75 * inch, 22, strokeColor=PRIMARY_COLOR, strokeWidth=2))
    story.append(d)
    story.append(Spacer(1, 10))
    
    # --- 2. Main Titles ---
    story.append(Paragraph("IF NOT, WHY NOT?", title_style))
    story.append(Spacer(1, 5))
    
    # --- 3. Introductory Hook ---
    intro_text = (
        "Do you know that Someone really cares about you? As a matter of fact, the price "
        "for your soul was His life, and that Someone is Jesus Christ. Jesus Christ is the Son of God "
    "who took the form of man so as to give Himself for man. The Bible tells us in <b>John 15:13</b>: "
        "<i>'Greater love hath no man than this, that a man lay down his life for his friends.'</i> "
        "The question to you today is: <b>have you made Him your friend? If not, why not?</b>"
    )
    story.append(Paragraph(intro_text, body_style))
    story.append(Paragraph("You can take these simple steps to make Him your friend:", body_style))
    
    # --- 4. The Action Steps (A-B-C-D) ---
    story.append(Paragraph("<b>A • Acknowledge</b> that you are a sinner <i>(Romans 3:23)</i>.", step_style))
    story.append(Paragraph("<b>B • Believe</b> that Jesus died to save you from sin <i>(Romans 5:8)</i>.", step_style))
    story.append(Paragraph("<b>C • Confess</b> your sins <i>(1 John 1:9)</i>.", step_style))
    story.append(Paragraph("<b>D • Depart</b> from sin and determine never to return to it <i>(2 Corinthians 6:17)</i>.", step_style))
    story.append(Spacer(1, 10))
    
    # --- 5. Styled Callout Box for the Sinner's Prayer ---
    prayer_content = [
        Paragraph("SINNER'S PRAYER", prayer_title),
        Paragraph(
            "\"Faithful Father, lover of my soul, I thank You for sending Jesus Christ Your only Son "
            "to give Himself for me. I believe that He came to bring me back to You and I surrender my sins "
            "today in exchange for a new life in You. I accept Him as my personal Lord and Saviour, thank "
            "You Father for forgiving me of my sins. This I pray in Jesus' name, Amen.\"", 
            prayer_body
        )
    ]
    
    container_width = 6.5 * inch
    prayer_table = Table([[prayer_content]], colWidths=[container_width])
    prayer_table.setStyle(TableStyle([
        ('BACKGROUND', (0,0), (-1,-1), BG_LIGHT),
        ('BOX', (0,0), (-1,-1), 1.5, PRIMARY_COLOR),
        ('TOPPADDING', (0,0), (-1,-1), 14),
        ('BOTTOMPADDING', (0,0), (-1,-1), 14),
        ('LEFTPADDING', (0,0), (-1,-1), 16),
        ('RIGHTPADDING', (0,0), (-1,-1), 16),
        ('ALIGN', (0,0), (-1,-1), 'CENTER'),
    ]))
    
    outer_table = Table([[prayer_table]], colWidths=[7.3 * inch])
    outer_table.setStyle(TableStyle([('ALIGN', (0,0), (-1,-1), 'CENTER')]))
    story.append(outer_table)
    
    # --- 6. Footer Affirmation ---
    story.append(Spacer(1, 10))
    story.append(Paragraph("Congratulations if you said this prayer of faith!<br/>Welcome to the family of God.", subtitle_style))
    story.append(Paragraph("God bless you.", footer_style))
    
    # Build document
    doc.build(story)

# Direct execution with zero matching syntax logic to clear out any typo errors
create_attractive_tract()
print("Beautifully styled tract layout generated successfully!")
```
