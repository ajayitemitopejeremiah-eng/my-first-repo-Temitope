```python
import urllib.request
import os

# Define the image assets and their corresponding target filenames
assets = {
    "praying_hands.png": "[attachment_3](attachment)",
    "vine_border.png": "[attachment_2](attachment)",
    "cross_glow.png": "[attachment_1](attachment)",
    "dove.png": "[attachment_0](attachment)"
}

print("Starting asset downloads...")

# Loop through and download each file safely
for filename, url in assets.items():
    try:
        print(f"Downloading {filename}...")
        # Include a user-agent header to ensure a smooth download process
        req = urllib.request.Request(url, headers={'User-Agent': 'Mozilla/5.0'})
        with urllib.request.urlopen(req) as response, open(filename, 'wb') as out_file:
            out_file.write(response.read())
        print(f"Successfully saved {filename}")
    except Exception as e:
        print(f"Failed to download {filename}. Error: {e}")

print("\nAll downloads finished. You can now execute your main image compilation script!")
```

```python
from PIL import Image

def generate_tract_design():
    # 1. Open the original base card image
    base_image = Image.open("base_card.jpeg").convert("RGBA")
    card_width, card_height = base_image.size

    # Helper function to open and resize transparent design elements
    def load_element(filename, target_width):
        img = Image.open(filename).convert("RGBA")
        # Calculate height keeping the aspect ratio perfect
        aspect_ratio = img.height / img.width
        target_height = int(target_width * aspect_ratio)
        return img.resize((target_width, target_height), Image.Resampling.LANCZOS)

    # 2. Load and resize each design asset
    # Adjust the target widths (pixels) to change asset sizes on the card
    hands = load_element("praying_hands.png", target_width=320)
    vine = load_element("vine_border.png", target_width=220)
    cross = load_element("cross_glow.png", target_width=180)
    dove = load_element("dove.png", target_width=140)

    # 3. Paste elements using coordinates matching your 2, 3, 4, 5 positions
    # Format: (X-coordinate, Y-coordinate). Top-left of the card is (0,0)

    # Position 2: Watermark style behind the Sinner's prayer box center
    # Adjust opacity of the hands to 35% so the text remains completely legible
    hands_alpha = hands.split()[3]
    hands_alpha = hands_alpha.point(lambda p: int(p * 0.35))
    hands.putalpha(hands_alpha)
    # Center the hands horizontally over the box area
    hands_x = (card_width - hands.width) // 2
    base_image.paste(hands, (hands_x, 340), mask=hands)

    # Position 3: Lower-left decorative border accent
    base_image.paste(vine, (20, card_height - vine.height - 40), mask=vine)

    # Position 4: Lower-right illuminated cross
    base_image.paste(cross, (card_width - cross.width - 40, card_height - cross.height - 40), mask=cross)

    # Position 5: Center bottom dove accent right next to the text
    # Placed right beside the "God bless you." text at the bottom center
    dove_x = (card_width // 2) + 60 
    base_image.paste(dove, (dove_x, card_height - dove.height - 35), mask=dove)

    # 4. Convert back to RGB format and save the final print-ready image
    final_design = base_image.convert("RGB")
    final_design.save("final_prayer_tract.jpg", "JPEG", quality=95)
    print("Design updated successfully! Checked final_prayer_tract.jpg")

if name == "main":
    generate_tract_design()
```
