```python
from scipy.stats import chi2_contingency
import pandas as pd

# 1. Define your target variable and a list of factors you want to test
target = 'practice_outcome'
factors = ['mother_age_cat', 'parity_cat','mother_hlfe', 'mother_occupationL', 'mother_marital_status', 'parents_religion', 'adequate_anc', 'place_of_birth', 'Previous_exp_source', 'sister_source', 'multiple_source', 'Grandmother_source', 'Other_Mothers_source', 'Friends_source', 'Healthcareworkers_source', 'Internet_source'] # Add all your column names here

# 2. Loop through each factor
for factor in factors:
    print(f"\n" + "="*30)
    print(f"ANALYSIS: {factor} vs {target}")
    print("="*30)
    
    # Create the frequency table (Counts)
    ct_counts = pd.crosstab(data[factor], data[target])
    
    # Create the percentage table (Row-wise percentages)
    ct_percent = pd.crosstab(data[factor], data[target], normalize='index') * 100
    
    # Perform the Chi-Square Test
    chi2, p, dof, expected = chi2_contingency(ct_counts)
    
    # Display Results
    print("--- Percentage Table (Row %) ---")
    print(ct_percent.round(2))
    
    print("\n--- Chi-Square Test Results ---")
    print(f"Chi-Square Statistic: {chi2:.4f}")
    print(f"P-value: {p:.4f}")
    print(f"Degrees of Freedom: {dof}")
    
    # Quick Interpretation
    alpha = 0.05
    if p < alpha:
        print("\nResult: Statistically significant (Reject H0)")
    else:
        print("\nResult: Not statistically significant (Fail to reject H0)")
```

```python
import pandas as pd
import statsmodels.api as sm
import numpy as np



# 2. Setup Variables
y = pd.factorize(data['practice_outcomeLR'])[0].astype(float)
covariates = ['mother_hlfeLR', 'parity_catLR', 'mother_age', 'adequate_ancLR', 'place_of_ancLR', 'mother_occupationLR', 'place_of_birthLR', 'residenceLR']

def run_model(cols):
    X = pd.get_dummies(data[cols], drop_first=True, dtype=int)
    X = sm.add_constant(X)
    # Using 'bfgs' to ensure convergence with the new categories
    model = sm.Logit(y, X).fit(method='bfgs', disp=False)
    return model

# 3. Build Table Rows
final_data = []

# Get Adjusted Results first (for all variables)
adj_model = run_model(covariates)
adj_ors = np.exp(adj_model.params)
adj_cis = np.exp(adj_model.conf_int())
adj_ps = adj_model.pvalues

# Loop through each variable to get Crude results
for var in covariates:
    crude_model = run_model([var])
    c_ors = np.exp(crude_model.params)
    c_cis = np.exp(crude_model.conf_int())
    c_ps = crude_model.pvalues
    
    for i in range(1, len(c_ors)):
        var_label = c_ors.index[i]
        
        # Format Crude
        crude_str = f"{c_ors[i]:.2f} ({c_cis.iloc[i,0]:.2f}-{c_cis.iloc[i,1]:.2f})"
        
        # Match with Adjusted values
        adj_str = f"{adj_ors[var_label]:.2f} ({adj_cis.loc[var_label,0]:.2f}-{adj_cis.loc[var_label,1]:.2f})"
        
        final_data.append({
            'Variable': var_label,
            'Crude OR (95% CI)': crude_str,
            'p-value (Crude)': f"{c_ps[i]:.3f}",
            'Adjusted OR (95% CI)': adj_str,
            'p-value (Adjusted)': f"{adj_ps[var_label]:.3f}"
        })

# 4. Display Final Table
final_table = pd.DataFrame(final_data)
print("--- FINAL PUBLICATION TABLE ---")
print(final_table.to_string(index=False))
```

```python
#for LR for safe cord care practice

import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
import numpy as np

# 1. Load Data
df = pd.read_excel("dataset_cord_care_analyze.xlsx")

# 2. Clean/Verify Data Types
df['practice_outcomeLR'] = pd.factorize(df['practice_outcomeLR'])[0]
df['mother_occupationL'] = df['mother_occupationL'].astype(str) # Force as text category

# 3. Define the 7-Predictor Formula
# Change 'Unemployed' to match the exact spelling of your baseline group in your Excel sheet
formula = (
    "mother_occupationL ~ C(mother_occupationL, Treatment('Unemployed')) + "
    "C(mother_hlfeLR) + C(parity_catLR) + C(adequate_ancLR) + C(place_of_ancLR)+ C(place_of_birthLR)+ C(residenceLR) + mother_age"
)

# 4. Fit Multivariable Logistic Regression
model_mv = smf.logit(formula, data=df).fit()

# 5. Extract Publication-Ready aOR Table
model_odds = np.exp(model_mv.params)
conf_int = np.exp(model_mv.conf_int())

df_aor = pd.DataFrame({
    'Adjusted Odds Ratio (aOR)': model_odds,
    'Lower CI (2.5%)': conf_int[0],
    'Upper CI (97.5%)': conf_int[1],
    'p-value': model_mv.pvalues
})

print(df_aor.round(3))
```

```python
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
import numpy as np

# 1. Load Data
df = pd.read_excel("dataset_cord_care_analyze.xlsx")

# 2. Clean/Verify Data Types & Handle Missing Values
df = df.dropna(subset=['practice_outcomeLR'])
df['practice_outcomeLR'] = df['practice_outcomeLR'].astype(int)
df['mother_occupationL'] = df['mother_occupationL'].astype(str)
df['mother_age'] = pd.to_numeric(df['mother_age'], errors='coerce')

# Define predictors exactly as they appear in your formula definitions
predictors = [
    "C(mother_occupationL, Treatment('Unemployed'))",
    "C(mother_hlfeLR)",
    "C(parity_catLR)",
    "C(adequate_ancLR)",
    "C(place_of_ancLR)",
    "C(place_of_birthLR)",
    "C(residenceLR)",
    "mother_age"
]

# --- STEP A: CALCULATE CRUDE ODDS RATIOS (UNIVARIATE) ---
crude_results = []

for pred in predictors:
    # Fit simple logistic regression for a single predictor
    uni_formula = f"practice_outcomeLR ~ {pred}"
    try:
        uni_model = smf.logit(uni_formula, data=df).fit(disp=0)
        
        # Extract odds, confidence intervals, and p-values
        or_val = np.exp(uni_model.params)
        ci_val = np.exp(uni_model.conf_int())
        p_val = uni_model.pvalues
        
        # Combine into a temporary dataframe for this predictor
        pred_df = pd.DataFrame({
            'cOR': or_val,
            'cOR_Lower': ci_val[0],
            'cOR_Upper': ci_val[1],
            'Crude p-value': p_val
        })
        crude_results.append(pred_df)
    except Exception as e:
        print(f"Skipping crude calculation for {pred} due to error: {e}")

# Combine all crude summaries together (excluding the individual intercept rows)
df_crude = pd.concat(crude_results)
df_crude = df_crude[~df_crude.index.str.contains('Intercept')]


# --- STEP B: CALCULATE ADJUSTED ODDS RATIOS (MULTIVARIABLE) ---
formula = "practice_outcomeLR ~ " + " + ".join(predictors)
model_mv = smf.logit(formula, data=df).fit()

model_odds = np.exp(model_mv.params)
conf_int = np.exp(model_mv.conf_int())

df_aor = pd.DataFrame({
    'aOR': model_odds,
    'aOR_Lower': conf_int[0],
    'aOR_Upper': conf_int[1],
    'Adjusted p-value': model_mv.pvalues
})


# --- STEP C: MERGE CRUDE AND ADJUSTED TABLES TOGETHER ---
# Join matching index strings to make sure categories line up side-by-side
df_final = df_aor.join(df_crude, how='left')

# Reorder columns logically for publication
final_column_order = [
    'cOR', 'cOR_Lower', 'cOR_Upper', 'Crude p-value',
    'aOR', 'aOR_Lower', 'aOR_Upper', 'Adjusted p-value'
]
df_final = df_final.reindex(columns=final_column_order)

print("\n" + "="*50)
print("FINAL MERGED LOGISTIC REGRESSION TABLE")
print("="*50)
print(df_final.round(3))
```
